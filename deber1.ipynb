```python
import numpy as np
import matplotlib.pyplot as plt
import random 
```

```python
def crea_malla(N, k,p):
    # diccionario guarda nodo como clave, conecciones como tupla dentro de lista
    # ejm= {(0,0): [(0,9),(0,1),(9,9)...]}
    lista_nodos = {} 
    # i j son el nodo 
    for i in range(N):
        for j in range(N):
            nodo = (i, j)
            vecinos = []
            #print(f"nodo {nodo}")
            for a in range(-k, k + 1): #a fila   ojo con k = 1 salen todas las direcciones necesarias sacado de chat gpt
                for b in range(-k, k + 1): # b columna
                    if a != 0 or b != 0:
                        if np.random.rand() <= p: # si es menor o igual a p se agrega un vecino aleatorio
                            #-----------preguntar profe si esta bien----------- 
                            nuevo_vecino = (np.random.randint(N) , np.random.randint(N)) #nodo aleatorio
                            while nuevo_vecino == nodo:# si el nodo aleatorio es igual al nodo actual se vuelve a generar
                               nuevo_vecino = (np.random.randint(N) , np.random.randint(N))
                            vecinos.append(nuevo_vecino)
                        else: # si no es menor o igual a p se agrega un vecino en la direccion a,b
                            nuevo_vecino = ((i + a) % N, (j + b) % N) # % N para que no se salga de la malla
                            #print(f"{a} {b} {nuevo_vecino}") 
                            vecinos.append(nuevo_vecino)
            lista_nodos[nodo] = vecinos
    return lista_nodos #diccionario con nodos y vecinos

def dibuja(lista_nodos):
    fig, ax = plt.subplots(figsize=(12,8))
    for nodo, vecinos in lista_nodos.items():
        for vecino in vecinos:
            ax.plot(*zip(nodo, vecino), color='lightblue')
        ax.scatter(*nodo, color='red')
        ax.text(*nodo, f"{nodo}", fontsize=14, ha='center', va='center')
    plt.axis('off')
    plt.show()
```

```python
N = 10 # tamaño de la malla
k = 1 # siempre dejar en 1, significa 8 conecciones 
p = 0.1 # probabilidad de que se conecte a un vecino aleatorio

lista_nodos = crea_malla(N, k,p)
dibuja(lista_nodos)
```

```python
#imprime lista nodos
for nodo, vecinos in lista_nodos.items():
    print(f"nodo {nodo}, conecciones {vecinos}")
```
